```python

# Charger les jeux de données
salary_data = pd.read_csv(r'C:\Users\Dell PC\Downloads\Dataset_labs\Salary_Data.csv')
insurance_data = pd.read_csv(r'C:\Users\Dell PC\Downloads\Dataset_labs\china_gdp.csv')

# Résumé statistique des deux jeux de données
print("====== Résumé statistique : Salary Data ======")
salary_stats = salary_data.describe()
print(salary_stats, "\n")

print("====== Résumé statistique : Insurance Data (China GDP) ======")
insurance_stats = insurance_data.describe()
print(insurance_stats, "\n")


```

```python
# Interprétation des résultats
print("=== Interprétation des Résumés Statistiques ===\n")

# Résumé et interprétation pour Salary Data
print("1. Salary Data:")
print(f"- Nombre d'observations : {salary_stats.loc['count', 'YearsExperience']} lignes.")
print(f"- Moyenne des années d'expérience : {salary_stats.loc['mean', 'YearsExperience']:.2f} ans.")
print(f"- Moyenne des salaires : {salary_stats.loc['mean', 'Salary']:.2f} unités monétaires.")
print(f"- L'expérience varie de {salary_stats.loc['min', 'YearsExperience']} à {salary_stats.loc['max', 'YearsExperience']} ans.")
print(f"- Les salaires varient de {salary_stats.loc['min', 'Salary']} à {salary_stats.loc['max', 'Salary']} unités monétaires.\n")

# Résumé et interprétation pour Insurance Data (China GDP)
print("2. Insurance Data (China GDP):")
print(f"- Nombre d'observations : {insurance_stats.loc['count', 'Year']} lignes.")
print(f"- La valeur moyenne du PIB : {insurance_stats.loc['mean', 'Value']:.2e} unités.")
print(f"- Le PIB varie de {insurance_stats.loc['min', 'Value']:.2e} à {insurance_stats.loc['max', 'Value']:.2e} unités.")
print(f"- Les données couvrent les années {insurance_stats.loc['min', 'Year']} à {insurance_stats.loc['max', 'Year']}.\n")
```
